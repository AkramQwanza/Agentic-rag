````python
import re
pattern = r'\[(\d+)\]'


# query ="les ventes trimestrielles d'une entreprise pour l'année 2023 sous forme de graphique en camembert (pie chart)"
# query ="Chiffre d'affaires trimestriel de l'entreprise A en 2023 pendant le 1ere  trimestre"
# query ="Chiffre d'affaires trimestriel de l'entreprise A en 2023 pendant le deuxième trimestre"
query = "comment est elle la Répartition des effectifs interne de la DSI par tranches d'âge"
answer = ex.run(query,k=3,similarity_score=0.6)
print("Question : "+ str(query))
print("Answer : "+ str(answer))

# Render Reference Slide
print("\n\nReference :")
matches = re.findall(pattern, answer)
for match in matches:
    print(ex.slides[int(match)-1].slide_text)
````
